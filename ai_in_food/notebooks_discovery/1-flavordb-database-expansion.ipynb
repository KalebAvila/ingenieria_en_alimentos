# Process the flavor db so it can be completely viewed:

## Imports

```python
import pandas as pd
```

# Configuration

```python
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.max_rows', 200)
```

## Read data

```python
flavor_db = pd.read_pickle('../data/external/flavor_db.pkl')
```

```python
flavordb_df = pd.DataFrame(flavor_db).T
```

```python
flavordb_df
```

```python
flavor_db_expanded = []
for entity in range(flavordb_df.shape[0]):
    entity_flavor_characteristics_df = pd.DataFrame(flavordb_df.drop(columns='molecules').iloc[entity]).T
    entity_flavor_characteristics_df['entity_index'] = entity
    entity_flavor_df = pd.DataFrame(flavordb_df.molecules.iloc[entity])
    entity_flavor_df['entity_index'] = entity
    complete_entity_flavor_df = entity_flavor_characteristics_df.merge(
        entity_flavor_df,
        on='entity_index', how='inner'
    ).drop(columns='entity_index')
    flavor_db_expanded.append(complete_entity_flavor_df)
```

```python
flavor_db_expanded_whole = pd.concat(flavor_db_expanded)
```

```python
flavor_db_expanded_whole
```

```python
flavor_db_expanded_whole.to_csv('../data/external/flavor_db_expanded.csv')
flavor_db_expanded_whole.to_pickle('../data/external/flavor_db_expanded.pkl')
```
