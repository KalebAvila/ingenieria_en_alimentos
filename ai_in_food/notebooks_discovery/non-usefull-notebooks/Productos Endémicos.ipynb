## Functions

```python
def read_data(path):
    return pd.read_csv(path, encoding='latin-1')
```

## Read data

```python
d1 = pd.read_csv(
    '../data_discovery/datos-abiertos/Cierre_agr_mun_2022.csv',
    encoding='latin-1'
)
d2 = pd.read_csv(
    '../data_discovery/datos-abiertos/Cierre_agricola_2002.csv',
    encoding='latin-1'
)
d3 = read_data(
    '../data_discovery/datos-abiertos/Cierre_agricola_mun_2021_no_seguimiento.csv'
)
d4 = read_data('../data_discovery/datos-abiertos/cierre_2005.csv')

d5 = read_data('../data_discovery/datos-abiertos/cierre_2022.csv')
```

```python
ingredientes_agricultura = list(
    list(d1.Nomcultivo.unique()) + 
    list(d2.Nomcultivo.unique()) + 
    list(d3.Nomcultivo.unique())
)

ingredients_agricultura = list(ingredientes_agricultura)

ingredientes_ganaderia = list(
    list((d5.Nomproducto + ' de ' + d5.Nomespecie).unique()) + 
    list((d4.Nomproducto + ' de ' + d4.Nomespecie).unique())
)
ingredientes = list(set(ingredients_agricultura + ingredientes_ganaderia))
```
