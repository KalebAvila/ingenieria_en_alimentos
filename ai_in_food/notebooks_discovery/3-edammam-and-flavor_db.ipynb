```python
found_in_edamam_wierd_lower = {}
for key in found_in_edamam_wierd.keys():
    new_key = key.lower().strip()
    found_in_edamam_wierd_lower[new_key] = found_in_edamam_wierd[key].lower().strip()


def changenames(string):
    string = string.lower().strip()
    if found_in_edamam_wierd_lower.get(string):
        return found_in_edamam_wierd_lower.get(string).replace(' ', '-')
    return string.lower().strip().replace(' ', '-')


flavor_db = pd.read_pickle('../data/external/flavor_db.pkl')
flavordb_df = pd.DataFrame(flavor_db).T

flavordb_df['search_names'] =  flavordb_df.entity_alias_readable.str.lower().apply(changenames)
```

```python
usefull_list = ['category', 'entity_id', 'entity_alias_readable', 'molecules', 'search_names']
edamam_flavor_db = flavordb_df[flavordb_df.search_names.isin(edammam_list)][usefull_list]
```

```python
edamam_flavor_db['json_path'] = '../data_discovery/edamam/edamam_' + edamam_flavor_db['search_names'] + '.json'
```

```python
def get_json_info(json_path):
    with open(json_path, 'r') as file: 
        f = json.load(file)
    return f

def get_similarity(x, ingredient):
    return SequenceMatcher(a=ingredient, b=x).ratio()

def food_ids(ingredient):
    hint_values = pd.DataFrame(
        pd.DataFrame(get_json_info(
            f"../data_discovery/edamam/edamam_{ingredient}.json"
        )['hints']).food.to_list()
    )
    step1 = hint_values[hint_values.label.str.lower().str.replace(' ','-') == ingredient]
    step2 = list(step1[step1.category=='Generic foods'].foodId.unique())
    if step2:
        return step2
    most_similar_ingredient = hint_values[
        hint_values.category=='Generic foods'
    ].label.apply(get_similarity, ingredient=ingredient) 
    if len(most_similar_ingredient):
        step3 = hint_values.loc[most_similar_ingredient.argmax()]
        return [step3.foodId]
    
    most_similar_ingredient2 = hint_values[(
        (hint_values.brand.isnull()) 
         & (hint_values.category=='Generic meals')
    )].label.apply(get_similarity, ingredient=ingredient)
    if len(most_similar_ingredient2):
        step4 = hint_values.loc[most_similar_ingredient2.argmax()]
        return [step4.foodId]
    
    most_similar_ingredient3 = hint_values.label.apply(get_similarity, ingredient=ingredient)
    if len(most_similar_ingredient3):
        step5 = hint_values.loc[most_similar_ingredient3.argmax()]
        return [step5.foodId]
    return []        
```

```python
edamam_flavor_db['possible_food_ids'] = edamam_flavor_db.search_names.apply(food_ids)
```
