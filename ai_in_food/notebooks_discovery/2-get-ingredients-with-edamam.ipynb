We are using the EDAMAM API with regars to the [Food DataBase API](https://developer.edamam.com/food-database-api-docs): 


This API grants us 100,000 calls per month in a free basis.
We can get the Food and nutrition Requests or UPC requests (35 per minute).


This database contains close to 900,000 common foods, restaurant menu items, grocery products and 615,000 unique UPCs/barcodes and 130,000 branded restaurant items. It also includes

- 10,000 generic foods	
- 90,000 generic meals	

The api has the following covered use cases: 
- Search for food by keyword, food name or UPC/Barcode
- Sourcing of nutrition facts for a given food, including: macro and micro nutrients, allergen labels, lifestyle and health labels
- Search for food by given nutrient quantity for 28 nutrients
- Search for foods within a given brand
- With the built-in food-logging context it allows for NLP requests for chatbots and natural language calorie counters

```python
def get_ingredient_info(ingredient):
    ingredient = ingredient.lower().replace(' ', '%20').strip()
    YOUR_APP_ID = '0abd8d1d'
    YOUR_APP_KEY = '1d961d26601c2a91f64b6e594255f8e2'
    base_url = 'https://api.edamam.com/api/food-database/v2/parser?'
    url = f'{base_url}app_id={YOUR_APP_ID}&app_key={YOUR_APP_KEY}&ingr={ingredient}&nutrition-type=cooking'
    info = requests.get(url)
    information = info.json()
    return information

def save_ingredient_info(information, ingredient):
    ingredient = ingredient.lower().replace(' ', '-')
    path = '../data_discovery/edamam/'
    with open(f'{path}edamam_{ingredient}.json', 'w') as f:
        json.dump(information, f)

def process_ingredient_data(ingredient):
    information = get_ingredient_info(ingredient)
    save_ingredient_info(information=information, ingredient=ingredient)
    return information
    
```
